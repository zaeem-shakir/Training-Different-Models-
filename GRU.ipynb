```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.datasets import fashion_mnist
from sklearn.metrics import accuracy_score, recall_score, precision_score

# Load Fashion MNIST dataset
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Reshape and normalize images
train_images = train_images.reshape(train_images.shape[0], 28, 28) / 255.0
test_images = test_images.reshape(test_images.shape[0], 28, 28) / 255.0

# Build and compile the GRU model
model = tf.keras.models.Sequential([
    layers.GRU(32, input_shape=(28, 28), return_sequences=True),
    layers.GRU(32),
    layers.Dropout(0.25),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.25),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['acc'])

# Train the model
model.fit(train_images, train_labels, batch_size=50, epochs=30)

# Predict and evaluate
predicted = model.predict(test_images)
predicted_labels = tf.argmax(predicted, axis=1)

# Calculate metrics
accuracy = accuracy_score(test_labels, predicted_labels)
precision = precision_score(test_labels, predicted_labels, average='weighted')
recall = recall_score(test_labels, predicted_labels, average='weighted')

print("Accuracy: {:.2%}".format(accuracy))
print("Precision: {:.2%}".format(precision))
print("Recall: {:.2%}".format(recall))
```
